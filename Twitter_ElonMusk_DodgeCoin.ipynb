```python
_2020 = pd.read_csv('2020.csv')
_2020.head()
```

```python
_2021 = pd.read_csv('2021.csv')
_2021.head()
```

```python
_2022 = pd.read_csv('2022.csv')
_2022.head()
```

```python
tweets = pd.concat([_2020, _2021, _2022], axis = 0)
tweets.head()
```

```python
tweets = tweets.reset_index(drop=True)
tweets
```

```python
tweets.info()
```

```python
tweets.describe()
```

```python
#only the tweets and the date
tweets = tweets[['date','tweet']]
tweets
```

```python
#clean up the date format
tweets.date = tweets.date.str[:10] #tweets.date.str[:10]
tweets
```

```python
#clean the tweets (remove @'s' and urls and)
TEXT_CLEANING_RE = "@\S+|https?:\S+|http?:\S|[^A-Za-z0-9]+"
new = []
for t in tweets.tweet:
    words = t.split()
    t = t.split()
    t = ' '.join(word for word in t if not word.startswith('@'))
    t = t.split()
    t = ' '.join(word for word in t if not word.startswith('http'))
    new.append(t)
    
tweets['tweet'] = new
tweets
```

```python
#find tweets mentioning dogecoin
doge = pd.DataFrame(columns = ['Date', 'Tweet'])

for i in range(len(tweets)):
    cur = tweets.loc[i].tweet
    cur = cur.lower()
    if 'doge' in cur:
        doge = doge.append({'Date' : tweets.loc[i].date, 'Tweet' : tweets.loc[i].tweet}, 
                ignore_index = True)

doge['Date'] = pd.to_datetime(doge['Date'], format='%Y-%m-%d')
doge
```
